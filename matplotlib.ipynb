**matplotlib** 

```python
#importing the matplotlib, pandas and numpy
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(y)
```

**Ploting the data**

```python
# ploting the sine wave
plt.plot(x,y)
plt.show()
```

```python
# build cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding the title x-axis and yaxis labels
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel('sine value')
plt.title("sine wave")
plt.show()
```

```python
# parabola
x = np.linspace(-10,10,20)
y = x**2
```

```python
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel('Parabolic value')
plt.title("parabolic wave")
plt.show()
```

```python
plt.plot(x,y, "r+")
plt.show()
```

```python
plt.plot(x,y, "r--")
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

```python
# bar plots
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lang = ["English", "french", "Spanish", "Latin", "German"]
people = [100, 50, 150, 40, 70]
ax.bar(lang, people)
plt.xlabel("Languages")
plt.ylabel('Number of people')
plt.title("bar Graph")
plt.show()
```

```python
# build the pie chart 

fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
lang = ["English", "french", "Spanish", "Latin", "German"]
people = [100, 50, 150, 40, 70]
ax.pie(people, labels=lang, autopct='%1.1f%%')

plt.title("Pir Chart")
```

```python
# Scatter plot

x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color="g")
ax.scatter(x,z,color="b")
plt.show()
```

```python
from matplotlib import projections
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z, c = z, cmap="Blues")
plt.show()
```
